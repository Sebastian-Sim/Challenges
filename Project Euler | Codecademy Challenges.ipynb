### Challenge 00. 

Write a program that prints the numbers from 1 to 100. But for multiples of three print Fizz instead of the number and for the multiples of five print Buzz. For numbers which are multiples of both three and five print FizzBuzz.

````python
all_num = range(1,101)

print("Three or Five Multiple")
for num in all_num:
    if num/15 == int(num/15):
        print(num, "- Fizz Buzz")
    elif num/3 == int(num/3):
        print(num, "- Fizz")
    elif num/5 == int(num/5):
        print(num, "- Buzz")
    else:
        print(num)
````

### Challenge 04.

Write a function that will take a given string and reverse the order of the words. “Hello world” becomes “world Hello” and “May the Fourth be with you” becomes “you with be Fourth the May”

````python
tip = input("Please provide sentence to be reverse: ")
tip_words = tip.split()
tip_words.reverse()

print("STRING:", tip)
print("LIST:", tip_words, "\n")

for word in tip_words:
    print(word)
````

````python
# [ ] review and run example
# triple quotes ''' ''' preserve formatting such as spaces and line breaks
big_quote = """Jack and Jill went up the hill
To fetch a pail of water
Jack fell down and broke his crown
And Jill came tumbling after"""

# split on line breaks (\n)
quote_lines = big_quote.split('\n')
print(quote_lines, '\n')

# print the list in reverse with index slicing
for line in quote_lines[::-1]:
    print(line)
````
